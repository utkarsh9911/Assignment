# Options data

```python
options_df = df[df["Ticker"].str.contains("BANKNIFTY") &
                (df["Ticker"].str.contains("CE") | df["Ticker"].str.contains("PE"))].copy()
```

```python
options_df.head()
```

```python
# 2. Extract expiry and scrip from Ticker
# Example Ticker: BANKNIFTY03FEB2235000PE.NFO
options_df["expiry"] = options_df["Ticker"].str.extract(r'BANKNIFTY(\d{2}[A-Z]{3}\d{2})')
options_df["scrip"] = options_df["Ticker"].str.extract(r'BANKNIFTY\d{2}[A-Z]{3}\d{2}(\d+CE|\d+PE)')
```

```python
options_df.head()
```

```python
# 3. Create datetime column (merge Date + Time)
options_df["datetime"] = pd.to_datetime(options_df["Date"] + " " + options_df["Time"])
```

```python
# Expiry format like 03FEB22  convert to datetime
options_df["expiry_date"] = pd.to_datetime(options_df["expiry"], format="%d%b%y")
```

```python
options_df["dte"] = (options_df["expiry_date"] - options_df["datetime"]).dt.days
```

```python
min_dte = options_df["dte"].min()
options_df = options_df[options_df["dte"] == min_dte]
```

```python
# 7. Drop unnecessary columns
options_df.drop(columns=["Ticker", "Date", "Time", "Volume", "Open Interest"], inplace=True)
```

```python
options_df["expiry"] = options_df["expiry_date"].dt.strftime("%d/%m/%Y")
```

```python
options_df.drop(columns=["expiry_date"], inplace=True)
```

```python
cols = ["datetime", "Open", "High", "Low", "Close", "expiry", "scrip", "dte"]
options_df = options_df[cols]
```

```python
# 8. Sort by datetime
options_df.sort_values(by="datetime", inplace=True)
options_df.reset_index(drop=True, inplace=True)
```

```python
options_df.head()
```

```python
options_df.to_csv("Options.csv", index=False)
```
